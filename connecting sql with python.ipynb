```python
# Que 9 What are seller types of swift cars?

car_km[car_km['Car_Name']=='swift']['Seller_Type'].unique()
```

```python
# Que 1 how many cars have price less than 100000?
num_cars = car_price[car_price['Selling_Price'] < 1.0].shape
print(f"Number of cars with price less than 100,000: {num_cars}")
```

```python
# # Que 2 Generate IDS C1 to C100 for all models and use it as index

ids = [f'C{i}' for i in range(1, 100)]

car_price.index=ids
car_price
##new_df_carkm=pd.DataFrame(car_price,index=ids)
##print(new_df_carkm)
```

```python
# Que 3 Create column difference in original and selling price

diff=car_price['Ori_Price']-car_price['Selling_Price']
diff
```

```python
# Que 4 Remove a column seller type in car_km datframe and add in car_price dataframe

#Remove 'seller_type' column from car_km and store it in a variable
seller_type=car_km.pop('Seller_Type')

## Add 'seller_type' column to car_price
car_price['Seller_Type']=seller_type
```

```python
print("Updated Car_km Dataframe")
print(car_km)
print("\nUpdated car_price Dataframe")
print(car_price)
```

```python
# Que 5 Which cars have automatic transmission

automatic_cars = car_km[car_km['Transmission'] == 'Automatic']
print(automatic_cars['Car_Name'])
```

```python
# Que 6 Create a dataframes of all Maruti and Toyota cars.


new_df=car_km[car_km['Car_Name'].isin(['fortuner','innova'])].reset_index()
del new_df['index']
new_df
```

```python
# Que 7 How many models are available for each car in database?

model_count=car_km.groupby('Car_Name')['Year'].nunique().reset_index()
model_count.columns=['Car_Name','model_count']
print(model_count)
```
